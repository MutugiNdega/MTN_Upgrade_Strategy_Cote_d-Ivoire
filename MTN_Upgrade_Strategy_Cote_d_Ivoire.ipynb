### 1.3 Cleaning Data

#### 1.3.1 Drop the useless and not needed columns

```python

Telcom_dataset = Telcom_dataset.drop(columns = ['CELL_ON_SITE' , 'COUNTRY_A' , 'COUNTRY_B'])
Telcom_dataset.columns
```

```python
Telcom_dataset2 = Telcom_dataset2.drop(columns = ['CELL_ON_SITE' , 'COUNTRY_A' , 'COUNTRY_B'])
Telcom_dataset2.columns
```

```python
Telcom_dataset3 = Telcom_dataset3.drop(columns = ['CELL_ON_SITE' , 'COUNTRY_A' , 'COUNTRY_B'])
Telcom_dataset3.columns
```

```python
cells_geo = cells_geo.drop(columns = 'Unnamed: 0')
cells_geo.columns
```

#### 1.3.2 Rename the misspelt columns

```python

Telcom_dataset = Telcom_dataset.rename(columns = {'PRODUTC':'PRODUCT' , 'DATETIME':'DATE_TIME'})
```

```python
Telcom_dataset2 = Telcom_dataset2.rename(columns = {'DW_A_NUMBER':'DW_A_NUMBER_INT' , 'DW_B_NUMBER':'DW_B_NUMBER_INT'})
```

```python
Telcom_dataset3 = Telcom_dataset3.rename(columns = {'SIET_ID':'SITE_ID' , 'CELLID':'CELL_ID'})
```

####1.3.3 Delete Duplicates

```python
Telcom_dataset.drop_duplicates()
```

```python
Telcom_dataset2.drop_duplicates()
```

```python
Telcom_dataset3.drop_duplicates()
```

```python
cells_geo.drop_duplicates()
```

#### 1.3.4 Dropping null values

```python
Telcom_dataset = Telcom_dataset.dropna(how = 'any')
Telcom_dataset
```

```python
Telcom_dataset2 = Telcom_dataset2.dropna(how = 'any')
Telcom_dataset2
```

```python
Telcom_dataset3 = Telcom_dataset3.dropna(how = 'any')
Telcom_dataset3
```

```python
cells_geo = cells_geo.dropna(how = 'any')
cells_geo
```

### 1.4 Merge the Tables

```python
Telcom = pd.concat([Telcom_dataset, Telcom_dataset2, Telcom_dataset3], axis=0)
Telcom
```

```python
MTN = pd.merge(Telcom, cells_geo, how='inner', left_on=['SITE_ID'], right_on=['SITE_CODE'])
MTN.head(5)
```

```python
#cities in high traffic zones
cities = zones['VILLES'].value_counts()
high_traffic_cities = cities[cities > 500]
high_traffic_cities
```

```python
cities_df = MTN[MTN['VILLES'].isin(high_traffic_cities.index)]
cities_df
```
